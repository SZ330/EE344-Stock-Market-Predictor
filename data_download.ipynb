```python
import yfinance as yf
import pandas as pd

df = yf.download(
    "SPY",
    start="1995-01-01",
    end="2025-12-31",
    auto_adjust=False,   # keep raw OHLC + Adj Close separately
    progress=False
)

df = df.reset_index()   # Date becomes a column
df.rename(columns={"Date":"Date"}, inplace=True)
df.to_csv("SPY_1995_2024_yfinance.csv", index=False)
print(df.head())
print(df.tail())
```
